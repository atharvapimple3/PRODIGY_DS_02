```python
import pandas as pd
df=pd.read_csv(r"D:/Downloads/titanic/train.csv",encoding='ISO-8859-1')
df.head()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Check for missing values
print(df.isnull().sum())

# Fill missing values (for simplicity, you can customize this based on your analysis)
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df['Fare'].fillna(df['Fare'].mean(), inplace=True)
```

```python
# Explore relationships between variables with pairplot
sns.pairplot(df, hue='Survived', height=2)
plt.title('Pairplot of Titanic Dataset')
plt.show()
```

```python
# Explore survival rate by different variables
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival Count by Gender')
plt.show()

```

```python
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title('Survival Count by Passenger Class')
plt.show()

```
